```python
import pandas as pd
from datetime import datetime as dt
```

```python
# Import the data and select the colums that might be useful
flights_df = pd.read_csv('data/flights.csv')
cols = ['YEAR','MONTH','DAY','DAY_OF_WEEK', 'FLIGHT_NUMBER', 'DESTINATION_AIRPORT','SCHEDULED_TIME', 'ELAPSED_TIME', 'WHEELS_ON','TAXI_IN','SCHEDULED_ARRIVAL','ARRIVAL_TIME','ARRIVAL_DELAY']
flights_df = flights_df[cols]
flights_df.head()
```

```python
# Create the datetime Column
date_cols = ["YEAR", "MONTH", "DAY"]
flights_df["DATETIME_OBJECT"] = pd.to_datetime(flights_df.loc[:,date_cols])
#Create the DATE_STRING column
flights_df['DATE_STRING'] = flights_df['DATETIME_OBJECT'].astype(str)
flights_df.head()
```

```python
# Reduce the dataframe to just the columns we want to use
calc_cols = ['ARRIVAL_DELAY', 'DESTINATION_AIRPORT', 'DATE_STRING']
flights_df = flights_df[calc_cols]
flights_df = flights_df.dropna()
flights_df.head()
```

```python
# Get the average delay per airpot per day:
average_delay_df = flights_df.groupby(['DATE_STRING', 'DESTINATION_AIRPORT']).mean()
average_delay_df.head()
```

```python
# Turn the groupby object into a dictionary:
avg_delay_dict = average_delay_df.to_dict()
avg_delay_dict
```
