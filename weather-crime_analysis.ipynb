```python
import pandas as pd
import datetime as dt
```

```python
# Import the Weather CSV as a Dataframe
max_temps = pd.read_csv("data/weather_data/daily_max_temps.csv")
# Convert the date column to datetime objects
max_temps.datetime = pd.to_datetime(max_temps.datetime)
# Add a column for the Year, Month, and Day of each row
max_temps['year'] = max_temps['datetime'].dt.year
max_temps['month'] = max_temps['datetime'].dt.month
max_temps['day'] = max_temps['datetime'].dt.day

max_temps.head()
```

```python
# Import the cleaned Crime Data CSV
crime_data = pd.read_csv("data/clean_crime_data.csv")
crime_cols = ['report_year', 'agency_jurisdiction', 'crimes_percapita']
crime_data = crime_data[crime_cols]
crime_data.head()
```
