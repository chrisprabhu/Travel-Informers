```python
import pandas as pd
from itertools import zip_longest
```

```python
# import the raw_crime_data CSV as a dataframe. 
crime_df = pd.read_csv("data/raw_crime_data.csv")
crime_df.head()
```

```python
# check that the report_year is a number and note a string
crime_df.dtypes
```

```python
len(crime_df)
```

```python
# choose the years you want to use
years_wanted = [2012, 2013, 2014, 2015]
```

```python
# Reduce the dataframe to include only the wanted years
crime_df = crime_df.loc[crime_df.report_year.isin(years_wanted)]
crime_df.head()
```

```python
# Cities included in the other dataset. 
project_cities_list = ["Vancouver","Portland","San Francisco","Seattle","Los Angeles","San Diego","Las Vegas","Phoenix","Albuquerque","Denver","SanAntonio","Dallas","Houston","Kansas City","Minneapolis","St. Louis","Chicago","Nashville","Indianapolis","Atlanta","Detroit","Jacksonville","Charlotte","Miami","Pittsburgh","Toronto","Philadelphia","New York","Montreal","Boston","Beersheba","Tel Aviv","Eilat","Haifa","Nahariyya","Jerusalem"]
project_cities_list = sorted(project_cities_list)
```

```python
# list of cities in the currently imported data set
datas_city_list = sorted(crime_df.agency_jurisdiction.tolist())
```

```python
cities_dict = {"Project Cities" : project_cities_list, 
               "Data's Cities": datas_city_list}
```

```python
for item in zip_longest(project_cities_list, datas_city_list, fillvalue='None'):
    print(item)
```

```python
pd.DataFrame(data=zip_longest(project_cities_list, datas_city_list, fillvalue='None'))
```

```python
# find the cities in the current data set which are in common with the other data set:
common_cities = []
for city in project_cities_list:
    for datas_city in datas_city_list:
        if city in datas_city:
            common_cities.append(datas_city)
common_cities
```

```python
# Reduce the common_cities list to a unique list of cities 
common_cities = pd.DataFrame(data=common_cities)[0].unique().tolist()
common_cities
```

```python
# Reduce dataframe to include only the cities that are common between both data sets. 
crime_df = crime_df.loc[crime_df.agency_jurisdiction.isin(common_cities)]
crime_df.head()
```

```python
#Save the new dataframe to a CSV "cleaned_data.csv"
crime_df.to_csv("data/cleaned_data.csv")
```
